```python
# Fourth Iteration
import sys
import json
from PyQt5.QtWidgets import (
    QApplication, QWidget, QVBoxLayout, QHBoxLayout,
    QLabel, QRadioButton, QPushButton, QDialog, QButtonGroup, QProgressBar
)
from PyQt5.QtGui import QFont
from PyQt5.QtCore import Qt, QTimer

class QuizApp(QWidget):
    def __init__(self):
        super().__init__()
        self.setWindowTitle("Simple Quiz Application")
        self.setGeometry(100, 100, 600, 400)
        self.current_question_index = 0
        self.score = 0
        self.total_time = 300  # Total quiz time in seconds (e.g., 5 minutes)
        self.remaining_time = self.total_time
        self.questions = self.load_questions("questions.json")
        self.init_ui()
        self.start_timer()

    def load_questions(self, file_path):
        try:
            with open(file_path, "r") as file:
                return json.load(file)
        except FileNotFoundError:
            self.show_error(f"File {file_path} not found!")
            sys.exit()
        except json.JSONDecodeError:
            self.show_error("Invalid JSON file format!")
            sys.exit()

    def init_ui(self):
        self.layout = QVBoxLayout()
        self.layout.setAlignment(Qt.AlignCenter)

        # Timer label
        self.timer_label = QLabel()
        self.timer_label.setFont(QFont("Arial", 14))
        self.timer_label.setAlignment(Qt.AlignRight)
        self.layout.addWidget(self.timer_label)

        # Question label
        self.question_label = QLabel()
        self.question_label.setFont(QFont("Arial", 14))
        self.question_label.setWordWrap(True)
        self.question_label.setAlignment(Qt.AlignCenter)
        self.layout.addWidget(self.question_label)

        # Progress bar
        self.progress_bar = QProgressBar()
        self.progress_bar.setMaximum(len(self.questions))
        self.layout.addWidget(self.progress_bar)

        # Options layout
        self.options_group = QButtonGroup(self)
        self.options_layout = QVBoxLayout()
        self.layout.addLayout(self.options_layout)

        # Submit button
        self.submit_button = QPushButton("Submit")
        self.submit_button.setFont(QFont("Arial", 12))
        self.submit_button.setStyleSheet("background-color: #4CAF50; color: white; padding: 10px;")
        self.submit_button.clicked.connect(self.check_answer)
        self.layout.addWidget(self.submit_button, alignment=Qt.AlignCenter)

        # Set main layout
        self.setLayout(self.layout)
        self.load_question()

    def start_timer(self):
        self.timer = QTimer(self)
        self.timer.timeout.connect(self.update_timer)
        self.timer.start(1000)  # 1-second intervals
        self.update_timer()

    def update_timer(self):
        if self.remaining_time > 0:
            self.remaining_time -= 1
            minutes, seconds = divmod(self.remaining_time, 60)
            self.timer_label.setText(f"Time Remaining: {minutes:02}:{seconds:02}")
        else:
            self.timer.stop()
            self.show_final_score()

    def load_question(self):
        if self.current_question_index < len(self.questions):
            question = self.questions[self.current_question_index]
            self.question_label.setText(f"Q{self.current_question_index + 1}: {question['question']}")
            self.clear_options()
            for i, option in enumerate(question["options"]):
                radio_button = QRadioButton(option)
                radio_button.setFont(QFont("Arial", 12))
                self.options_group.addButton(radio_button, id=i)
                self.options_layout.addWidget(radio_button)

            # Update progress bar
            self.progress_bar.setValue(self.current_question_index + 1)
        else:
            self.show_final_score()

    def clear_options(self):
        for button in self.options_group.buttons():
            self.options_group.removeButton(button)
            button.deleteLater()

    def check_answer(self):
        selected_id = self.options_group.checkedId()
        if selected_id == -1:
            self.show_error("Please select an option!")
            return

        correct_answer = self.questions[self.current_question_index]["answer"]
        if selected_id == correct_answer:
            self.score += 1

        self.current_question_index += 1
        self.load_question()

    def show_final_score(self):
        self.timer.stop()
        elapsed_time = self.total_time - self.remaining_time
        minutes, seconds = divmod(elapsed_time, 60)
        dialog = QDialog(self)
        dialog.setWindowTitle("Quiz Completed")
        dialog.setGeometry(150, 150, 400, 250)

        # Layout for dialog
        layout = QVBoxLayout()
        layout.setAlignment(Qt.AlignCenter)

        # Title
        title = QLabel("Congratulations!")
        title.setFont(QFont("Arial", 16, QFont.Bold))
        title.setAlignment(Qt.AlignCenter)
        layout.addWidget(title)

        # Determine message based on score
        total_questions = len(self.questions)
        if self.score == total_questions:
            message = "Outstanding! You aced the quiz!"
            color = "#4CAF50"  # Green
        elif self.score >= total_questions * 0.7:
            message = "Great job! You did really well."
            color = "#FFBF00"  # Yellow/Gold
        elif self.score >= total_questions * 0.4:
            message = "Good effort! Keep practicing to improve."
            color = "#FF8000"  # Orange
        else:
            message = "Don't give up! Try again to improve your score."
            color = "#FF0000"  # Red

        # Score display
        score_label = QLabel(f"Your Score: {self.score}/{total_questions}")
        score_label.setFont(QFont("Arial", 14))
        score_label.setStyleSheet(f"color: {color};")
        score_label.setAlignment(Qt.AlignCenter)
        layout.addWidget(score_label)

        # Time display
        time_label = QLabel(f"Time Taken: {minutes:02}:{seconds:02}")
        time_label.setFont(QFont("Arial", 12))
        time_label.setAlignment(Qt.AlignCenter)
        layout.addWidget(time_label)

        # Custom message
        message_label = QLabel(message)
        message_label.setFont(QFont("Arial", 12))
        message_label.setAlignment(Qt.AlignCenter)
        layout.addWidget(message_label)

        # Close button
        close_button = QPushButton("Close")
        close_button.setFont(QFont("Arial", 12))
        close_button.setStyleSheet("background-color: #FF5733; color: white; padding: 10px;")
        close_button.clicked.connect(dialog.close)
        layout.addWidget(close_button, alignment=Qt.AlignCenter)

        dialog.setLayout(layout)
        dialog.exec_()

    def show_error(self, message):
        dialog = QDialog(self)
        dialog.setWindowTitle("Error")
        dialog.setGeometry(150, 150, 400, 150)

        # Layout for dialog
        layout = QVBoxLayout()
        layout.setAlignment(Qt.AlignCenter)

        # Error message
        error_label = QLabel(message)
        error_label.setFont(QFont("Arial", 12))
        error_label.setStyleSheet("color: red;")
        error_label.setAlignment(Qt.AlignCenter)
        layout.addWidget(error_label)

        # Close button
        close_button = QPushButton("OK")
        close_button.setFont(QFont("Arial", 12))
        close_button.setStyleSheet("background-color: #FF5733; color: white; padding: 10px;")
        close_button.clicked.connect(dialog.close)
        layout.addWidget(close_button, alignment=Qt.AlignCenter)

        dialog.setLayout(layout)
        dialog.exec_()

if __name__ == "__main__":
    app = QApplication(sys.argv)
    window = QuizApp()
    window.show()
    sys.exit(app.exec_())
```
